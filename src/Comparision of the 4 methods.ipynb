```python
import numpy as np
import matplotlib.pyplot as plt

error_responses_1 = np.load('error_responses_CGA.npy')
error_responses_2 = np.load('error_responses_CBA.npy')
error_responses_3 = np.load('error_responses_EGA.npy')
error_responses_4 = np.load('error_responses_EBA.npy')

# Frequency array (assumed to be the same for all methods)
frequencies = np.logspace(-5, 5, num=40)

# Collect all error responses in a list for plotting
errors = [error_responses_1, error_responses_2, error_responses_3, error_responses_4]
labels = ['CGA', 'CBA', 'EGA', 'EBA']
colors = ['b', 'g', 'r', 'c']
#linestyles = ['-', '--', '-.', ':']
#markers = ['o', '^', 's', 'x']

# Create a plot to compare the errors
plt.figure(figsize=(10, 6))
for i, error in enumerate(errors):
    plt.loglog(frequencies, error, color=colors[i], label=labels[i])

plt.xlabel('Frequency (rad/s)')
plt.ylabel('Error (H∞ Norm Difference)')
plt.title('Comparison of Errors Between Four Reduction Methods')
plt.legend()
plt.grid(True, which='both', ls='--')
plt.tight_layout()
plt.savefig('C:/Users/G-shop/Desktop/RAPPORT-Master/comparision_norm_plot2.png')
plt.show()
```
